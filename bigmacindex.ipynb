> ### 1. מהו עקרון שוויון כוח הקנייה
![](https://raw.githubusercontent.com/nivlevi1/Bigmac-s-Economy/main/text1.PNG)
![](https://raw.githubusercontent.com/nivlevi1/Bigmac-s-Economy/main/text2.PNG)

```python
url = 'https://raw.githubusercontent.com/TheEconomist/big-mac-data/master/output-data/big-mac-full-index.csv'
mc_df = pd.read_csv(url)
mc_df.head()
```

```python
mc_df['year'] = pd.DatetimeIndex(mc_df.date).year
```

---
> ### 2. האם אפשר לקנות המבורגר בפחות משעת עבודה ?
![](https://raw.githubusercontent.com/nivlevi1/Bigmac-s-Economy/main/text3.PNG)

```python
data_isr= mc_df.loc[(mc_df.iso_a3 == 'ISR')]
data_isr.head()
```

```python
isr_plot= data_isr.groupby(['year'], as_index=False)['local_price'].aggregate('max')
isr_plot['MIN_hourly_wage'] = [15.94, 17.56, 20.70, 20.70, 22.04, 23.12, 23.12, 23.12, 25.00, 25.94, 28.49, 29.12, 29.12, 29.12, 29.12, 29.66]
#credit https://www.btl.gov.il/Mediniyut/GeneralData/Pages/%D7%A9%D7%9B%D7%A8%20%D7%9E%D7%99%D7%A0%D7%99%D7%9E%D7%95%D7%9D.aspx
```

```python
plot = isr_plot.plot(x="year", y=["local_price",'MIN_hourly_wage'], figsize=(12, 6), color=['red','gold'], marker='o')
plt.title('BigMac Price VS MIN hourly wage in Israel', fontsize=14)
plt.xlabel('Year', fontsize=14)
plt.ylabel('NIS (₪)', fontsize=14)
plt.show()
```

---
> ### 3. בחזרה למדד הביג מאק
![](https://raw.githubusercontent.com/nivlevi1/Bigmac-s-Economy/main/text4.PNG)

```python
p22 =  mc_df.loc[(mc_df.year == 2022)]
p22g = p22.groupby(['name'], as_index=False)['dollar_price'].aggregate('mean').sort_values(['dollar_price'],ascending=False)
plt.figure(figsize=(12, 10))
sns.barplot(x=p22g["dollar_price"], y=p22g['name'] ,orient = "h", edgecolor=".3", palette="YlOrBr")
plt.title('BigMac Price in each country', fontsize=14)
plt.xlabel('Price ($)', fontsize=10)
plt.ylabel('Country', fontsize=10)
plt.show()
```

---
> ### 4. מדד מבוסס שערי המרה
![](https://raw.githubusercontent.com/nivlevi1/Bigmac-s-Economy/main/RAW-index%20-%20ECONOMIST%20credit.PNG)
![](https://raw.githubusercontent.com/nivlevi1/Bigmac-s-Economy/main/text5.PNG)

```python
index22 =  mc_df.loc[(mc_df.year == 2022)]
index22 = index22.loc[:,['name','dollar_price', 'adj_price', 'GDP_dollar','USD_raw','USD_adjusted','USD_adjusted']].sort_values(['USD_raw'],ascending=False)
index22.reset_index(inplace = True, drop = True)
index22.iloc[[0,1,3,-3,-2,-1]]
```

```python
g= sns.relplot(x="name", y="USD_raw",
            sizes=(40, 400), alpha=.5, palette="muted",
            height=6, data=index22)
plt.xticks(rotation=90)
g.fig.set_size_inches(10,5)
plt.axhline(y=0, color='tomato', linestyle='-')
plt.text(0, 0.01, "Overvalued", horizontalalignment='left', size='large', color='green', weight='semibold')
plt.text(0, -0.04, "Undervalued", horizontalalignment='left', size='large', color='red', weight='semibold')
plt.title('Raw index, relative to the US dollar', fontsize=14)
plt.xlabel('Country', fontsize=14)
plt.ylabel('USD RAW ($)', fontsize=14)
plt.show()
```
